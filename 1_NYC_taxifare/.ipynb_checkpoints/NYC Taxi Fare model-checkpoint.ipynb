```python
import os
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
root = r'D:/data/1_NYC_city_taxi/preprocessed/'
train_fname = "hav_clust_date.csv"
test_fname = "hav_clust_date_test.csv"

```

```python
df_train = pd.read_csv(os.path.join(root,train_fname))
```

```python
df_train.head()
```

```python
columns = df_train.columns
```

```python
pick_up_columns = list(columns)[2:]
```

```python
pick_up_columns.remove('fare_amount')
```

```python
pick_up_columns.remove('pickup_datetime')
```

```python
df_train[pick_up_columns]
```

```python



x_train, x_test , y_train, y_test =  train_test_split(df_train[pick_up_columns],df_train['haversine'], 
                                                     test_size = 0.3 ,
                                                     random_state=123)
```

```python
import time
```

```python
time.time()
```

```python
# Scaling before model fit
from sklearn.preprocessing import StandardScaler

start_time = time.time()
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

rf = RandomForestRegressor(n_estimators=300, random_state=123456)
rf.fit(x_train, y_train)
end_time = time.time()


print("elapse time", end_time - start_time)
```

```python

### testing n_job option

start_time = time.time()
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

rf = RandomForestRegressor(n_estimators=300, random_state=123456,n_jobs=8)

rf.fit(x_train, y_train)
end_time = time.time()
print("elapse time", end_time - start_time)
```

```python

### testing n_job option

start_time = time.time()
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

rf = RandomForestRegressor(n_estimators=300, random_state=123456,n_jobs=-1)

rf.fit(x_train, y_train)
end_time = time.time()
print("elapse time", end_time - start_time)
```

```python

start_time = time.time()
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

rf = RandomForestRegressor(n_estimators=400, random_state=123456,n_jobs=8)

rf.fit(x_train, y_train)
end_time = time.time()
print("elapse time", end_time - start_time)
```

```python
from sklearn import metrics

y_pred = rf.predict(x_test)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred))) 
```

## Hyper Parameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 50, stop = 150, num = 20)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 80, num = 10)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

#
rfc = RandomForestRegressor()

# 
rf_random = RandomizedSearchCV(estimator = rfc, param_distributions = random_grid, n_iter = 3,
                               cv = 3, verbose=2, random_state=42, n_jobs = 8)

rf_random.fit(x_train,y_train)

```

```python
best_random = rf_random.best_estimator_
```

```python
from sklearn import metrics

y_pred = best_random.predict(x_test)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred))) 
```
